## Cleaning Pipeline

### ID Column Cleaning
Ensures all ID columns are strings, stripped of spaces, and consistently formatted

```python
id_columns = ["session_id", "viewer_id", "streamer_id", "stream_id"]

def clean_ids(df):
    df = df.copy()
    for col in id_columns:
        if col in df.columns:
            df[col] = df[col].astype(str).str.strip()
    return df
```

```python
datasets = [clean_ids(df) for df in datasets]

df1, df2, df3, df4 = datasets
```

### Convert Date Columns

```python
def clean_dates(df):
    df = df.copy()
    for col in ["started_at", "ended_at"]:
        if col in df.columns:
            df[col] = pd.to_datetime(df[col], errors="coerce")
    return df

df1 = clean_dates(df1)
df3 = clean_dates(df3)
```

### => Validate Date Conversion Cleaning
Try to catch weird rows where recorded duration doesn't match timestamps

```python
# Sessions Calculations
df1["calculated_duration_mins"] = (
    df1["ended_at"] - df1["started_at"]
).dt.total_seconds() / 60

df1.head(3)
```

```python
# Streams Calculations
df3["calculated_stream_duration_hrs"] = (
    df3["ended_at"] - df3["started_at"]
).dt.total_seconds() / 3600

df3.head(3)
```

## Check Duplicate Entries

```python
def check_key_duplicates(df, key, name):
    duplicates = df.duplicated(subset=key).sum()
    print(f"{name}: {duplicates} duplicate rows by {key}")

check_key_duplicates(df1, ["session_id"], "Dataset 1")
check_key_duplicates(df2, ["streamer_id"], "Dataset 2")
check_key_duplicates(df3, ["stream_id", "streamer_id"], "Dataset 3") # Check if one steam id can appear under mult streamers
check_key_duplicates(df4, ["viewer_id"], "Dataset 4")
```

## Clean Categorical Columns

```python
def clean_categories(df):
    df = df.copy()

    categorical_cols = [
        "category",
        "language",
        "age_group",
        "country",
        "subscription_tier",
        "preferred_category"
    ]

    for col in categorical_cols:
        if col in df.columns:
            df[col] = df[col].astype(str).str.strip()

    return df

df3 = clean_categories(df3)
df2 = clean_categories(df2)
df4 = clean_categories(df4)
```

## Validate Numeric Columns

```python
numeric_checks = {
    "duration_mins": df1,
    "chat_messages_sent": df1,
    "bits_cheered": df1,
    "total_followers": df2,
    "avg_concurrent_viewers": df2,
    "stream_duration_hrs": df3,
    "peak_viewers": df3,
    "account_age_days": df4,
}

def validate(checks):
    for col, df in checks.items():
        print(col)
        print("Missing:", df[col].isna().sum())
        print("Negative:", (df[col] < 0).sum())
        print()
```

```python
validate(numeric_checks)
```

## Cleaning Numeric Columns (Drop Entries)

#### Helper functions for cleaning

```python
def drop_invalid_numeric_rows(df, numeric_cols):
    df = df.copy()

    valid_mask = True

    for col in numeric_cols:
        if col in df.columns:
            valid_mask = valid_mask & df[col].notna() & (df[col] >= 0)

    return df[valid_mask]

# Sets invalid values to missing instead of total drop
def replace_invalid_numeric_with_na(df, numeric_cols):
    df = df.copy()

    for col in numeric_cols:
        if col in df.columns:
            df.loc[df[col] < 0, col] = pd.NA

    return df
```

```python


# Fill missing count-like activity values with 0
df1["chat_messages_sent"] = df1["chat_messages_sent"].fillna(0)
df1["bits_cheered"] = df1["bits_cheered"].fillna(0)

df2["total_followers"] = df2["total_followers"].fillna(0)
df2["avg"] = df2["total_followers"].fillna(0)



# Drop impossible values
df1 = drop_invalid_numeric_rows(
    df1,
    ["duration_mins", "chat_messages_sent", "bits_cheered"]
)

df2 = replace_invalid_numeric_with_na(
    df2,
    ["total_followers", "avg_concurrent_viewers"]
)

df3 = replace_invalid_numeric_with_na(
    df3,
    ["stream_duration_hrs", "peak_viewers"]
)

df4 = drop_invalid_numeric_rows(
    df4,
    ["account_age_days"]
)

og_dataset = datasets

# Update the dataset with cleaned versions
datasets = [df1, df2, df3, df4]

# Preserve Removed Rows
df2_removed = og_dataset[1].loc[~og_dataset[1].index.isin(datasets[1].index)]
df3_removed = og_dataset[2].loc[~og_dataset[2].index.isin(datasets[2].index)]
```

### Check if Removed IDs appear in Df1

If an entry has been drop, if affects the number of sessions.

```python
removed_streamers_used = df1[df1["streamer_id"].isin(df2_removed["streamer_id"])]

print("Sessions affected by removed streamers:", len(removed_streamers_used))
print(removed_streamers_used["streamer_id"].value_counts())
```

#### Streams (Effect)

```python
removed_streams_used = df1[df1["stream_id"].isin(df3_removed["stream_id"])]

print("Sessions affected by removed streams:", len(removed_streams_used))
print(removed_streams_used["stream_id"].value_counts().head(20))
```

## Merge Dataset for Exploration

Category inquiry: 
Both Df2 and Df3 have a "category" column.


My proposed approach/assumption:<br>
Dataset 2’s category means the streamer’s usual/current category.<br>
Dataset 3’s category means the category of that specific stream.<br>

```python
df2 = df2.rename(columns={
    "category": "streamer_category"
})

df3 = df3.rename(columns={
    "category": "stream_category",
    "started_at": "stream_started_at",
    "ended_at": "stream_ended_at"
})
```

Main Activity/session table; using df1 as base

```python
merged = (
    df1
    .merge(df2, on="streamer_id", how="left", validate="many_to_one")
    .merge(df4, on="viewer_id", how="left", validate="many_to_one")
    .merge(df3, on=["stream_id", "streamer_id"], how="left", validate="many_to_one")
)
```
