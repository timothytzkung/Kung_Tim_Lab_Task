### Grading-focused relationship and type audit

The grading rubric asks for both data types and relationships, so I treat `sessions` as the central fact table and the other three tables as dimension tables. The next cells create a compact data dictionary and verify whether the foreign keys in `sessions` match the available profile/stream metadata.

```python
table_map = {
    "sessions": df1,
    "streamers": df2,
    "streams": df3,
    "viewers": df4,
}

# Data dictionary: one row per column, including dtype, unique values, missing values, and example values.
data_dictionary = []
for table_name, table in table_map.items():
    for col in table.columns:
        data_dictionary.append({
            "table": table_name,
            "column": col,
            "dtype": str(table[col].dtype),
            "non_null": table[col].notna().sum(),
            "missing": table[col].isna().sum(),
            "unique_values": table[col].nunique(dropna=True),
            "sample_values": table[col].dropna().astype(str).head(3).tolist(),
        })

data_dictionary = pd.DataFrame(data_dictionary)
display(data_dictionary)
```

### Task 1 Summary Report

The four files form a relational dataset. `sessions` is the central activity table because each row represents a viewer's interaction with a stream and contains the key IDs needed to join the other tables. `viewers` contains viewer-level demographics and subscription information, `streamers` contains streamer-level profile attributes, and `streams` contains stream-event metadata such as stream category, timing, duration, and peak viewers.

The main one-to-many relationships are `viewer_id`, `streamer_id`, and `stream_id`: one viewer can appear in many sessions, one streamer can host many streams/sessions, and one stream can have many viewer sessions. The relationship audit above is important because it identifies whether any session records lack matching profile or stream metadata. Those unmatched rows should be retained for analyses that only need session-level variables, but filtered or flagged for analyses that rely on missing stream/streamer attributes.

## General/All: Cleaning Pipeline

### ID Column Cleaning
Ensures all ID columns are strings, stripped of spaces, and consistently formatted

```python
id_columns = ["session_id", "viewer_id", "streamer_id", "stream_id", "subscription_tier"]

def clean_ids(df):
    df = df.copy()
    for col in id_columns:
        if col in df.columns:
            df[col] = df[col].astype(str).str.strip()
    return df
```

### DF4 (Viewers) Column Cleaning

```python
datasets = [clean_ids(df) for df in datasets]

df1, df2, df3, df4 = datasets
```

```python
# Make a cleaned copy of the column
df4["age_group"] = (
    df4["age_group"]
    .astype(str)
    .str.strip()
    .str.replace("–", "-", regex=False)      # replace en dash with normal hyphen
    .str.replace(" to ", "-", regex=False)   # replace "18 to 24" with "18-24"
)

# Check result
df4["age_group"].value_counts()
```

#### => Convert Age Group To Categorical Variable

```python
age_order = ["13-17", "18-24", "25-34", "35-44", "45+"]

df4["age_group"] = pd.Categorical(
    df4["age_group"],
    categories=age_order,
    ordered=True
)
# Check result
df4["age_group"].value_counts().sort_index()
```

#### => Normalize Subscription Tier

```python
# Normalize subscription tier
df4["subscription_tier"] = (
    df4["subscription_tier"]
    .astype(str)
    .str.strip()
    .str.lower()
    .str.replace("premium", "tier1", regex=False)
    .str.replace("free trial", "free", regex=False)
    .str.replace("none", "free", regex=False)
)
```

```python
# Convert to categorical
tier_order = ["free", "tier1", "tier2", "tier3", "tier4"]
df4["subscription_tier"] = pd.Categorical(
    df4["subscription_tier"],
    categories=tier_order,
    ordered=True
)
```

```python
# validate cleaned set
df4["subscription_tier"].value_counts()
```

## Temporal Data Cleaning

### DF1 & DF3: Convert Date Columns

```python
def clean_dates(df):
    df = df.copy()
    for col in ["started_at", "ended_at"]:
        if col in df.columns:
            df[col] = pd.to_datetime(df[col], errors="coerce")
    return df

df1 = clean_dates(df1)
df3 = clean_dates(df3)
```

#### DF1: Calculate True Duration
Try to catch weird rows where recorded duration doesn't match timestamps

```python
# Sessions Calculations
df1["calculated_duration_mins"] = (
    df1["ended_at"] - df1["started_at"]
).dt.total_seconds() / 60

df1.head(3)
```

#### DF1: => Perform Column Swap (Dates)
Performs column swap on ended_at and started_at stamps if calculated_duration_mins is the invert of duration_mins

```python
# In place columns swap
_df1 = df1[
    (df1["duration_mins"] == -1 * df1["calculated_duration_mins"]) & (df1["duration_mins"] > 0)
]
_df1["started_at"], _df1["ended_at"] =  _df1["ended_at"].copy(), _df1["started_at"].copy()

_df1["calculated_duration_mins"] = (
    _df1["ended_at"] - _df1["started_at"]
).dt.total_seconds() / 60

_df1.shape
```

```python
# Drop uncleaned valued from df1 & then merge
df1 = (
    pd.concat([df1, _df1], ignore_index=True)
    .drop_duplicates(subset=["session_id"], keep="last")
    .reset_index(drop=True)
)
df1.shape
```

#### DF1: => Update Duration Mins

```python
df1["duration_mins"] = df1["calculated_duration_mins"]
```

#### => Validate Date Conversion Comparisons

##### DF1 - Sessions

```python
session_duration_mismatches = df1[
    df1["duration_mins"] != df1["calculated_duration_mins"]
]

print("Session duration mismatches:", len(session_duration_mismatches))
session_duration_mismatches
```

## DF3 Cleaning

```python
# Streams Calculations
df3["calculated_stream_duration_hrs"] = (
    df3["ended_at"] - df3["started_at"]
).dt.total_seconds() / 3600

df3.head(3)
```

### => Update Duration Hours

```python
_df3 = df3[
    df3["stream_duration_hrs"] != df3["calculated_stream_duration_hrs"]
]
_df3["stream_duration_hrs"] = _df3["calculated_stream_duration_hrs"]

# Drop uncleaned values
df3 = df3[
    df3["stream_duration_hrs"] == df3["calculated_stream_duration_hrs"]
]
df3 = pd.concat([df3, _df3], ignore_index=True)
df3
```

### Validate Duration Hours

```python
def check_key_duplicates(df, key, name):
    duplicates = df.duplicated(subset=key).sum()
    print(f"{name}: {duplicates} duplicate rows by {key}")

check_key_duplicates(df1, ["session_id"], "Dataset 1")
check_key_duplicates(df2, ["streamer_id"], "Dataset 2")
check_key_duplicates(df3, ["stream_id", "streamer_id"], "Dataset 3") # Check if one stream id can appear under mult streamers
check_key_duplicates(df4, ["viewer_id"], "Dataset 4")
```

## All: Clean Categorical Columns

```python
def clean_categories(df):
    df = df.copy()

    categorical_cols = [
        "category",
        "language",
        "age_group",
        "country",
        "subscription_tier",
        "preferred_category"
    ]

    for col in categorical_cols:
        if col in df.columns:
            df[col] = df[col].astype(str).str.strip()

    return df

df3 = clean_categories(df3)
df2 = clean_categories(df2)
df4 = clean_categories(df4)
```

### Task 2 summary report

The cleaning process focuses on preserving the relational structure of the four datasets while correcting values that would distort analysis. ID columns were standardized as strings so that joins are not affected by whitespace or accidental numeric formatting. Age groups were normalized into a consistent ordered set, and subscription tiers were normalized so that equivalent labels such as missing/free-style values do not split the same viewer segment into multiple categories.

For time variables, the notebook recalculates duration from timestamps and uses this derived duration as the more reliable value when the recorded duration conflicts with the start/end times. This is justified because duration should be mechanically determined by the timestamps. Negative numeric values are treated as invalid because counts and durations cannot be below zero. Rather than dropping entire rows, the affected cells are converted to missing values so the session, viewer, stream, or streamer can still contribute to analyses that do not depend on that specific field.

The merged dataset is built from `sessions` as the base table. Unmatched stream or streamer metadata is reported instead of hidden. This matters because missing metadata after a join is a data coverage issue, not necessarily a bad session record.

### Descriptive statistics by viewer, time, and stream attributes

These summaries move beyond overall averages by comparing watch behaviour across viewer demographics, subscription tiers, stream categories, and time of day.

```python
analysis_df = merged_all.copy()

# Prefer session start time for time-of-day analysis. Coerce protects the notebook if the column is still object dtype.
analysis_df["started_at"] = pd.to_datetime(analysis_df["started_at"], errors="coerce")
analysis_df["session_hour"] = analysis_df["started_at"].dt.hour
analysis_df["daypart"] = pd.cut(
    analysis_df["session_hour"],
    bins=[-1, 5, 11, 17, 23],
    labels=["overnight", "morning", "afternoon", "evening"]
)

summary_overall = analysis_df[[
    "duration_mins", "chat_messages_sent", "bits_cheered", "peak_viewers",
    "total_followers", "avg_concurrent_viewers", "account_age_days"
]].describe().T

display(summary_overall)
```

```python
viewer_time_summary = (
    analysis_df
    .groupby(["age_group", "subscription_tier"], observed=True)
    .agg(
        sessions=("session_id", "count"),
        median_watch_mins=("duration_mins", "median"),
        mean_watch_mins=("duration_mins", "mean"),
        median_chat_messages=("chat_messages_sent", "median"),
        median_bits=("bits_cheered", "median"),
    )
    .reset_index()
    .sort_values(["age_group", "subscription_tier"])
)

display(viewer_time_summary)
```

```python
stream_category_summary = (
    analysis_df
    .dropna(subset=["stream_category"])
    .groupby("stream_category")
    .agg(
        sessions=("session_id", "count"),
        median_watch_mins=("duration_mins", "median"),
        mean_watch_mins=("duration_mins", "mean"),
        median_peak_viewers=("peak_viewers", "median"),
        median_chat_messages=("chat_messages_sent", "median"),
    )
    .sort_values("sessions", ascending=False)
)

display(stream_category_summary.head(10))
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
sns.boxplot(data=analysis_df, x="daypart", y="duration_mins", ax=ax)
ax.set_title("Session Duration by Time of Day")
ax.set_xlabel("Time of day")
ax.set_ylabel("Session duration (minutes)")
plt.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(12, 6))
top_categories = analysis_df["stream_category"].value_counts().head(8).index
sns.barplot(
    data=analysis_df[analysis_df["stream_category"].isin(top_categories)],
    x="stream_category",
    y="duration_mins",
    estimator="median",
    errorbar=None,
    ax=ax,
)
ax.set_title("Median Watch Time for Top Stream Categories")
ax.set_xlabel("Stream category")
ax.set_ylabel("Median session duration (minutes)")
ax.tick_params(axis="x", rotation=45)
plt.tight_layout()
plt.show()
```

## All: Cleaning Numeric Columns (Replace Entries)

#### Helper functions for cleaning

```python
def drop_invalid_numeric_rows(df, numeric_cols):
    df = df.copy()

    valid_mask = True

    for col in numeric_cols:
        if col in df.columns:
            valid_mask = valid_mask & df[col].notna() & (df[col] >= 0)

    return df[valid_mask]

# Sets invalid values to missing instead of total drop
def replace_invalid_numeric_with_na(df, numeric_cols):
    df = df.copy()

    for col in numeric_cols:
        if col in df.columns:
            df.loc[df[col] < 0, col] = pd.NA

    return df
```

### Replace Missing Values with Either "Missing" or 0

```python
# Preserve pre-cleaning versions so we can compare the effect of each strategy.
og_dataset = [df1.copy(), df2.copy(), df3.copy(), df4.copy()]

# Count-like activity columns: missing usually means no recorded activity, so 0 is a defensible imputation.
df1["chat_messages_sent"] = df1["chat_messages_sent"].fillna(0)
df1["bits_cheered"] = df1["bits_cheered"].fillna(0)

# For scale variables, negative values are impossible but should not cause entire ID rows to be removed.
# I convert only the impossible cells to NA so joins and relationships remain intact.
df1 = replace_invalid_numeric_with_na(df1, ["duration_mins", "chat_messages_sent", "bits_cheered"])
df2 = replace_invalid_numeric_with_na(df2, ["total_followers", "avg_concurrent_viewers"])
df3 = replace_invalid_numeric_with_na(df3, ["stream_duration_hrs", "peak_viewers"])
df4 = replace_invalid_numeric_with_na(df4, ["account_age_days"])

# Update the dataset list with cleaned versions.
datasets = [df1, df2, df3, df4]

# Track whether any rows would have been lost under a destructive drop strategy.
df2_removed = og_dataset[1].loc[~og_dataset[1].index.isin(datasets[1].index)]
df3_removed = og_dataset[2].loc[~og_dataset[2].index.isin(datasets[2].index)]
```

### Missing-value and outlier diagnostics

For Task 4, I separate missing-value detection from outlier detection. Missing values affect whether a row can be used for a specific analysis, while outliers may represent either errors or valid extreme behaviour.

```python
def missing_report(df, name):
    report = pd.DataFrame({
        "column": df.columns,
        "missing_count": df.isna().sum().values,
        "missing_percent": (df.isna().mean().values * 100).round(2),
    })
    report.insert(0, "dataset", name)
    return report.sort_values("missing_percent", ascending=False)

missing_reports = pd.concat([
    missing_report(df1, "sessions"),
    missing_report(df2, "streamers"),
    missing_report(df3, "streams"),
    missing_report(df4, "viewers"),
], ignore_index=True)

display(missing_reports[missing_reports["missing_count"] > 0])
```

```python
def iqr_outlier_report(df, name, numeric_cols):
    rows = []
    for col in numeric_cols:
        if col not in df.columns:
            continue
        s = pd.to_numeric(df[col], errors="coerce").dropna()
        if s.empty:
            continue
        q1, q3 = s.quantile([0.25, 0.75])
        iqr = q3 - q1
        lower = q1 - 1.5 * iqr
        upper = q3 + 1.5 * iqr
        rows.append({
            "dataset": name,
            "column": col,
            "q1": q1,
            "q3": q3,
            "iqr": iqr,
            "lower_fence": lower,
            "upper_fence": upper,
            "outlier_count": ((s < lower) | (s > upper)).sum(),
            "outlier_percent": (((s < lower) | (s > upper)).mean() * 100).round(2),
        })
    return pd.DataFrame(rows)

outlier_reports = pd.concat([
    iqr_outlier_report(df1, "sessions", ["duration_mins", "chat_messages_sent", "bits_cheered"]),
    iqr_outlier_report(df2, "streamers", ["total_followers", "avg_concurrent_viewers"]),
    iqr_outlier_report(df3, "streams", ["stream_duration_hrs", "peak_viewers"]),
    iqr_outlier_report(df4, "viewers", ["account_age_days"]),
], ignore_index=True)

display(outlier_reports.sort_values("outlier_percent", ascending=False))
```

```python
# Compare two missing-value strategies on selected analysis columns.
strategy_cols = ["duration_mins", "chat_messages_sent", "bits_cheered", "peak_viewers", "account_age_days"]
strategy_cols = [c for c in strategy_cols if c in merged.columns]

comparison_rows = []
for col in strategy_cols:
    original = pd.to_numeric(merged[col], errors="coerce")
    median_imputed = original.fillna(original.median())
    zero_imputed = original.fillna(0)
    dropped = original.dropna()

    comparison_rows.extend([
        {"column": col, "strategy": "drop_missing", "n_used": len(dropped), "mean": dropped.mean(), "median": dropped.median()},
        {"column": col, "strategy": "median_impute", "n_used": len(median_imputed), "mean": median_imputed.mean(), "median": median_imputed.median()},
        {"column": col, "strategy": "zero_impute", "n_used": len(zero_imputed), "mean": zero_imputed.mean(), "median": zero_imputed.median()},
    ])

imputation_comparison = pd.DataFrame(comparison_rows)
display(imputation_comparison)
```

```python
def cap_iqr(series):
    s = pd.to_numeric(series, errors="coerce")
    q1, q3 = s.quantile([0.25, 0.75])
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    return s.clip(lower, upper)

# Compare two outlier strategies using duration as a concrete example.
duration = pd.to_numeric(merged["duration_mins"], errors="coerce")
q1, q3 = duration.quantile([0.25, 0.75])
iqr = q3 - q1
lower, upper = q1 - 1.5 * iqr, q3 + 1.5 * iqr

outlier_strategy_comparison = pd.DataFrame([
    {"strategy": "original", "n_used": duration.notna().sum(), "mean": duration.mean(), "median": duration.median(), "max": duration.max()},
    {"strategy": "remove_iqr_outliers", "n_used": duration.between(lower, upper).sum(), "mean": duration[duration.between(lower, upper)].mean(), "median": duration[duration.between(lower, upper)].median(), "max": duration[duration.between(lower, upper)].max()},
    {"strategy": "cap_iqr_outliers", "n_used": duration.notna().sum(), "mean": cap_iqr(duration).mean(), "median": cap_iqr(duration).median(), "max": cap_iqr(duration).max()},
])

display(outlier_strategy_comparison)
```

**Interpretation:** Median imputation is generally safer for skewed behavioural variables because it is less influenced by extreme sessions. Zero imputation is appropriate only when missingness plausibly means no activity, such as missing chat or bits fields. For outliers, capping keeps all rows while reducing the influence of extremes; removing outliers creates a cleaner distribution but can also delete legitimate heavy users.

### Task 4 Summary report

The missing-value audit identifies which columns lose usable information and whether missingness appears in the original tables or after merging. The outlier audit uses the IQR rule to flag unusually high or low values in numeric columns. These flags are not treated as automatic deletion rules because extreme viewer behaviour may be real, especially for variables such as chat messages, bits cheered, peak viewers, and session duration.

Two missing-value strategies are compared: dropping missing rows and imputing missing values. Median imputation is preferred for skewed continuous variables because it preserves row count without being overly affected by extreme values. Zero imputation is used only for count-like engagement variables where missing can reasonably mean no recorded activity. Two outlier strategies are also compared: removing IQR outliers and capping them at the IQR fences. Capping is usually safer for this dataset because it retains unusual but potentially valid high-engagement sessions while limiting their effect on averages and correlations.

### Viewer-level check-ins and correlation analysis

Correlation should be calculated at the correct unit of analysis. Since the task asks about the number of check-ins, I aggregate sessions to the viewer level before correlating viewer features with session counts.

```python
viewer_behavior = (
    merged_all
    .groupby("viewer_id")
    .agg(
        check_ins=("session_id", "count"),
        avg_duration_mins=("duration_mins", "mean"),
        total_watch_mins=("duration_mins", "sum"),
        total_chat_messages=("chat_messages_sent", "sum"),
        total_bits_cheered=("bits_cheered", "sum"),
        account_age_days=("account_age_days", "first"),
        age_group=("age_group", "first"),
        subscription_tier=("subscription_tier", "first"),
    )
    .reset_index()
)

# Encode ordered categories numerically for correlation. These codes are ordinal, so Spearman is often more appropriate.
viewer_behavior["age_group_code"] = pd.Categorical(viewer_behavior["age_group"], categories=age_order, ordered=True).codes
viewer_behavior["subscription_tier_code"] = pd.Categorical(viewer_behavior["subscription_tier"], categories=tier_order, ordered=True).codes

corr_cols = [
    "check_ins", "avg_duration_mins", "total_watch_mins", "total_chat_messages",
    "total_bits_cheered", "account_age_days", "age_group_code", "subscription_tier_code"
]

pearson_corr = viewer_behavior[corr_cols].corr(method="pearson")
spearman_corr = viewer_behavior[corr_cols].corr(method="spearman")

display(pearson_corr.round(3))
display(spearman_corr.round(3))
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(pearson_corr, annot=True, cmap="coolwarm", center=0, vmin=-1, vmax=1, fmt=".2f", ax=ax)
ax.set_title("Pearson Correlation Matrix: Viewer-Level Behaviour")
plt.tight_layout()
plt.show()

fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(spearman_corr, annot=True, cmap="coolwarm", center=0, vmin=-1, vmax=1, fmt=".2f", ax=ax)
ax.set_title("Spearman Correlation Matrix: Viewer-Level Behaviour")
plt.tight_layout()
plt.show()
```

```python
target = "check_ins"
correlation_summary = pd.DataFrame({
    "pearson_with_check_ins": pearson_corr[target].drop(target),
    "spearman_with_check_ins": spearman_corr[target].drop(target),
}).sort_values("spearman_with_check_ins", key=lambda s: s.abs(), ascending=False)

display(correlation_summary.round(3))
```

**Interpretation:** Pearson correlation measures linear relationships and is most appropriate when variables are continuous and roughly symmetric. Spearman correlation uses ranks, so it is better for skewed behavioural variables, ordinal categories, and monotonic relationships that are not perfectly linear. In this dataset, Spearman is especially useful for variables such as subscription tier, age group, total chat messages, and total bits, because they are ordered or likely skewed.

## Student-defined Question 1: Do viewers watch longer when the stream category matches their preferred category?

This question connects viewer preference data with actual session behaviour. It is stronger than only counting popular categories because it tests whether preference alignment is associated with longer engagement.

```python
preference_df = merged_all.dropna(subset=["preferred_category", "stream_category", "duration_mins"]).copy()
preference_df["preference_match"] = (
    preference_df["preferred_category"].astype(str).str.lower().str.strip()
    == preference_df["stream_category"].astype(str).str.lower().str.strip()
)

preference_summary = (
    preference_df
    .groupby("preference_match")
    .agg(
        sessions=("session_id", "count"),
        unique_viewers=("viewer_id", "nunique"),
        median_duration=("duration_mins", "median"),
        mean_duration=("duration_mins", "mean"),
        median_chat=("chat_messages_sent", "median"),
    )
    .reset_index()
)

display(preference_summary)

fig, ax = plt.subplots(figsize=(8, 5))
sns.barplot(data=preference_df, x="preference_match", y="duration_mins", estimator="median", errorbar=None, ax=ax)
ax.set_title("Median Watch Time by Preference Match")
ax.set_xlabel("Stream category matches viewer preferred category")
ax.set_ylabel("Median session duration (minutes)")
plt.tight_layout()
plt.show()
```
